# Part1 単一の記事の取得

ここでは小売・流通業のメディアであるDiamond Chain Store( "https://diamond-rm.net/" )を使って演習をします。サイトから記事を１つ選んでください。

解答例は "https://diamond-rm.net/market/87934/" の記事になっていますが、ご自身が選択された記事のURLに読み替えてください。
また今後、記事が非公開になる可能性があります。予めご了承ください。

### タスク1： newspaperからArticleをインポートしてください。

### タスク2： Articleを使って記事のダウンロードと解析を行い、記事の発行日を表示してください。

※記事によっては発行日が設定されていない場合があります。

### タスク3： 記事の著者を表示してください。

※記事によっては著者が設定されていない場合があります。

### タスク4： 記事の本文を表示してください。

# Part2 複数の記事の取得

### タスク2:  ITMedia( https://www.itmedia.co.jp/ )、THE BRIDGE( https://thebridge.jp/ )のサイトのトップページに表示されている記事をそれぞれ順に10本取得し、タイトル、URLと記事の内容を表示してください。
